```python
print("Task 7")
sub1 = float(input("Enter marks 1: "))
sub2 = float(input("Enter marks 2: "))
sub3 = float(input("Enter marks 3: "))
sub4 = float(input("Enter marks 4: "))
sub5 = float(input("Enter marks 5: "))

total = sub1 + sub2 + sub3 + sub4 + sub5
percentage = total / 5

print("Total:", total)
print("Percentage:", percentage)

if percentage >= 80:
    print("Grade A")
elif percentage >= 70:
    print("Grade B")
elif percentage >= 60:
    print("Grade C")
elif percentage >= 50:
    print("Grade D")
else:
    print("Grade F")
```
